# Experimenting with deep learning

Let the model classify aquarium fishes or an environmental sounds


```
from IPython.display import display
from IPython.display import Audio
from fastai.vision.widgets import widgets
from fastaudio.ci import skip_if_ci
from ipywidgets import HBox, Layout, VBox
from fastai.vision.all import *
from fastaudio.all import AudioTensor

lbl_env = widgets.Label('environmental sound')
lbl_aqu = widgets.Label('aquarium fish')

out_pl_env = widgets.Output()
out_pl_aqu = widgets.Output()

btn_upload_env = widgets.FileUpload(accept='.wav')
btn_upload_aqu = widgets.FileUpload(accept='.jpg')

lbl_pred_env = widgets.Label()
lbl_pred_aqu = widgets.Label()

def dump_to_file(content):
    fname = 'tmp.wav'
    f = open(fname, 'wb')
    f.write(content)
    f.close()
    return fname

def on_data_change_env(change):
    lbl_pred_env.value = ''
    out_pl_env.clear_output()
    content = btn_upload_env.data[-1]
    fname = dump_to_file(content)
    learn_inf = load_learner('environmental_sound_model.pkl', cpu=True)
    audio = AudioTensor.create(fname)
    with out_pl_env: display(Audio(content))
    pred,pred_idx,probs = learn_inf.predict(audio)
    lbl_pred_env.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

def on_data_change_aqu(change):
    lbl_pred_aqu.value = ''
    out_pl_aqu.clear_output()
    img = PILImage.create(btn_upload_aqu.data[-1])
    with out_pl_aqu: display(img.to_thumb(128,128))  
    learn_inf = load_learner('aquarium_fish_model.pkl', cpu=True)  
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred_aqu.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_upload_env.observe(on_data_change_env, names=['data'],)
btn_upload_aqu.observe(on_data_change_aqu, names=['data'])

box_env= HBox([lbl_env, btn_upload_env, out_pl_env, lbl_pred_env],layout=Layout(
    display='flex',
    flex_flow='row',
    border='4mm ridge rgba(170, 50, 220, .6)',
    align_items='stretch',
    width='70%'
))
box_aqu= HBox([lbl_aqu, btn_upload_aqu, out_pl_aqu, lbl_pred_aqu],layout=Layout(
    display='flex',
    flex_flow='row',
    border='4mm ridge rgba(170, 50, 220, .6)',
    align_items='stretch',
    width='70%'
))

display(VBox([box_env, box_aqu]))
```
